# K means Clustering (Unsupervised learning)

1. Select value of k (no. of clusters)


2. Initialize centroids randomly.


3. Measure the euclidian distance bw the data points and tries to gather the data points which are closer to the centroid.


4. The points above & below the circle are collected into seprate clusters.


5. It will calculate the mean of all datapoints. It will be called as centroid.


6. Once the centroid is calculated, the randomly selecte centroid shifts to the newly collected mean value.


7. The centroid again tries to build its cluster and at this point there could be movement of points from one cluster to another.


8. The above steps (3-7) keeps on repeating until there is no movement of points from one cluster to another and two clusters are distinctly figured out.

The right value to choose k value is doing by elbow curve. 
T a certain k value the Accuracy stops increasing and error not decreasing. By identifying that k value it is selected.
